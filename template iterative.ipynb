```python
import matplotlib.pyplot as plt
def dibujarTray(tray):
    xs = [float(vec[0]) for vec in tray]
    ys = [float(vec[1]) for vec in tray]

    plt.plot(xs, ys, marker="o")
    plt.title("Trayectoria")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.show()
dibujarTray(tray)
```
