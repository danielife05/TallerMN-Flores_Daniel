```python
import matplotlib.pyplot as plt

def dibujarTray(tray1, tray2):
    xs1 = [float(vec[0]) for vec in tray1]
    ys1 = [float(vec[1]) for vec in tray1]

    xs2 = [float(vec[0]) for vec in tray2]
    ys2 = [float(vec[1]) for vec in tray2]

    plt.plot(xs1, ys1, marker="o", label="Gauss Jacobi")
    plt.plot(xs2, ys2, marker="o", label="Gauss Seidel")
    plt.title("Trayectorias")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.legend()
    plt.show()

def dibujarSoloTray(tray1):
    xs1 = [float(vec[0]) for vec in tray1]
    ys1 = [float(vec[1]) for vec in tray1]

    plt.plot(xs1, ys1, marker="o", label="Trayectoria")
    plt.title("Trayectoria")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.legend()
    plt.show()

# No converge
x, tray = gauss_jacobi(A=[[2,10],[3,2]], b=[16,11], x0=[1, 1], tol=0.0001, max_iter=200)
x2, tray2 = gauss_seidel(A=[[2,10],[3,2]], b=[16,11], x0=[1, 1], tol=0.0001, max_iter=2000)
dibujarTray(tray, tray2)
# Gauss-Jacobi
dibujarSoloTray(tray)
# Gauss-Seidel
dibujarSoloTray(tray2)
```

```python
# Converge con x0 = [1, 1]
x, tray = gauss_jacobi(A=[[3,2],[2,10]], b=[11,16], x0=[1, 1], tol=0.0001, max_iter=200)
x2, tray2 = gauss_seidel(A=[[3,2],[2,10]], b=[11,16], x0=[1, 1], tol=0.0001, max_iter=200)
dibujarTray(tray, tray2)
```
